# 整理df1

```python
df1['稻作期數'] = df1.apply(lambda x:1 if x['作物名稱']=='一期稻作' else (2 if x['作物名稱']=='二期稻作' else '無資料'),axis=1)
```

```python
df1['颱風日期'] = df1.apply(lambda x:datetime.datetime.strptime(str(x['年度']+1911)+str(x['災害起始日'])[-4:], "%Y%m%d").date(),axis=1)
```

```python
df1['插植起始日_起'] = df1.apply(lambda x:x['插植時期'].split('-')[0],axis=1)
df1['插植起始日_末'] = df1.apply(lambda x:x['插植時期'].split('-')[1],axis=1)
df1['插植起始日_起'] = df1.apply(lambda x:re.sub('月|日','',x['插植起始日_起']),axis=1)
df1['插植起始日_末'] = df1.apply(lambda x:re.sub('月|日','',x['插植起始日_末']),axis=1)
```

```python
def date_convert(date,year):
    if len(date)==2:
        return datetime.datetime.strptime(str(year+1911)+'0'+date[0:1]+'0'+date[-1:], "%Y%m%d").date()
    elif len(date)==3:
        return datetime.datetime.strptime(str(year+1911)+'0'+date[0:1]+date[1:], "%Y%m%d").date()
    elif len(date)==4:
        return datetime.datetime.strptime(str(year+1911)+date, "%Y%m%d").date()    
    else:
        return ''
    
def check(start,end):
    try:
        if start<end:
            return 1
        else:
            return 0
    except:
        return 0
    
def date_correct(date,check,year):
    if check==0:   
        if len(date)==4:
            return datetime.datetime.strptime(str(year+1910)+date, "%Y%m%d").date()
        elif len(date)==3 and date[0:2]=='10':
            return datetime.datetime.strptime(str(year+1911)+date[0:2]+'0'+date[-1:], "%Y%m%d").date()
        else:
            return date_convert(date,year)
    else:
        return date_convert(date,year)  
```

```python
df1['插植起始日_起(date)'] = df1.apply(lambda x:date_convert(x['插植起始日_起'],x['年度']),axis=1)
df1['插植起始日_末(date)'] = df1.apply(lambda x:date_convert(x['插植起始日_末'],x['年度']),axis=1)
```

```python
df1['check'] = df1.apply(lambda x:check(x['插植起始日_起(date)'],x['插植起始日_末(date)']),axis=1)
```

```python
df1[(df1['check']==0)&(df1['插植起始日_起']!='')&(df1['插植起始日_末']!='')]
```

```python
df1['插植起始日_起(date)'] = df1.apply(lambda x:date_correct(x['插植起始日_起'],x['check'],x['年度']),axis=1)
df1['插植起始日_末(date)'] = df1.apply(lambda x:date_correct(x['插植起始日_末'],x['check'],x['年度']),axis=1)
df1['check'] = df1.apply(lambda x:check(x['插植起始日_起(date)'],x['插植起始日_末(date)']),axis=1)
```

```python
df1[(df1['check']==0)&(df1['插植起始日_起']!='')&(df1['插植起始日_末']!='')]
```
