### 爬稻米資料

```python
options = Options()
options.add_argument("--disable-gpu");
chrome = webdriver.Chrome('./chromedriver',options=options)
#稻米
chrome.get("https://agr.afa.gov.tw/afa/pgricetown_cond.jsp")
```

```python
#設定等待時間
chrome.implicitly_wait(5)
chrome.set_page_load_timeout(30)
```

```python
#稻米按鍵設定
def year_button(i):
    select = Select(chrome.find_element_by_name('accountingyear'))
    select.select_by_value(i)

def item_button(i):
    select = Select(chrome.find_element_by_name('item'))
    select.select_by_value(i)

def crop_button(i):
    select = Select(chrome.find_element_by_name('crop'))
    select.select_by_value(i)

def city_button(i):
    select = Select(chrome.find_element_by_name('city'))
    select.select_by_value(i)
    
def send(chrome):
    send = chrome.find_element_by_class_name('InputBtn')
    send.click()
```

### 建立抓取資料迴圈

```python
#稻米按建
year_list = ['098','099','100','101','102','103','104','105','106','107','108','109']
item_list = ['01','02']
crop_list = ['C01','C02','C03','C04','C05','C06']
city_list = ['0001','0002','0003','0004','0005','0006','0007','0008','0009','0010','0011','0012','0013','0014','0015','0016','0017','0018','0020','0063','0065','0066']
```

```python
#稻米建立4種dataframe(實際種植面積,收穫面積,稻穀總產量,糙米總產量) 分年,期作,稻種存取留於後續合併
for year in year_list:
    for item in item_list:
        for crop in crop_list:
            df_node = pd.DataFrame()
            for county in city_list:
                try:
                    year_button(year)
                    item_button(item)
                    crop_button(crop)
                    city_button(county)
                    send(chrome)
                    #抓取表格內容建立dataframe
                    table_tr_list = chrome.find_element_by_name('afa').find_elements_by_tag_name('tr')
                    table_list = []
                    for tr in table_tr_list:
                        #將每一個tr的資料根據td查詢出來，返回結果為list物件
                        table_td_list = tr.find_elements_by_tag_name('td')
                        row_list = []
                        for td in table_td_list:    #遍歷每一個td
                            row_list.append(td.text)    #取出表格的資料，並放入行列表裡
                        table_list.append(row_list)
                    if table_list[-1][0] in ['合\u3000\u3000計','查無資料！！']:
                        df = pd.DataFrame(table_list[2:-1],columns = table_list[0])
                    else:
                        df = pd.DataFrame(table_list[2:],columns = table_list[0])
                    df_node = pd.concat([df_node,df])
                    print('完成{}年第{}期作{}稻種{}......前往下一步!'.format(year,item,crop,county))
                    time.sleep(2)
                    chrome.back()
                except TimeoutException as e:
                    print('在{}年第{}期作{}稻種{}載入時發生問題......跳過往下一步走!'.format(year,item,crop,county))
                    chrome.back()
            globals()['df_area_{}_{}_{}'.format(year,item,crop)] = df_node[['縣市鄉鎮名稱','實際種植面積']]
            globals()['df_harvest_{}_{}_{}'.format(year,item,crop)] = df_node[['縣市鄉鎮名稱','收穫面積']]
            globals()['df_yield_{}_{}_{}'.format(year,item,crop)] = df_node[['縣市鄉鎮名稱','稻穀總產量']]
            globals()['df_chao_{}_{}_{}'.format(year,item,crop)] = df_node[['縣市鄉鎮名稱','糙米總產量']]
```

```python
#稻米先存檔
for year in year_list:
    for item in item_list:
        for crop in crop_list:
            globals()['df_area_{}_{}_{}'.format(year,item,crop)].to_excel(u'D:\\Terry\\工作文件\\氣象局計畫\\資料\\稻米災損資料\\rawdata\\df_area_{}_{}_{}.xlsx'.format(year,item,crop),index=False)
            globals()['df_harvest_{}_{}_{}'.format(year,item,crop)].to_excel(u'D:\\Terry\\工作文件\\氣象局計畫\\資料\\稻米災損資料\\rawdata\\df_harvest_{}_{}_{}.xlsx'.format(year,item,crop),index=False)
            globals()['df_yield_{}_{}_{}'.format(year,item,crop)].to_excel(u'D:\\Terry\\工作文件\\氣象局計畫\\資料\\稻米災損資料\\rawdata\\df_yield_{}_{}_{}.xlsx'.format(year,item,crop),index=False)
            globals()['df_chao_{}_{}_{}'.format(year,item,crop)].to_excel(u'D:\\Terry\\工作文件\\氣象局計畫\\資料\\稻米災損資料\\rawdata\\df_chao_{}_{}_{}.xlsx'.format(year,item,crop),index=False)
```

```python
#將稻米儲存資料轉為數字
for year in year_list:
    for item in item_list:
        for crop in crop_list:
            globals()['df_area_{}_{}_{}'.format(year,item,crop)]['實際種植面積']=globals()['df_area_{}_{}_{}'.format(year,item,crop)]['實際種植面積'].map(lambda x:re.sub(',','',x))
            globals()['df_area_{}_{}_{}'.format(year,item,crop)]['實際種植面積']=globals()['df_area_{}_{}_{}'.format(year,item,crop)]['實際種植面積'].astype(float)
            globals()['df_harvest_{}_{}_{}'.format(year,item,crop)]['收穫面積']=globals()['df_harvest_{}_{}_{}'.format(year,item,crop)]['收穫面積'].map(lambda x:re.sub(',','',x))
            globals()['df_harvest_{}_{}_{}'.format(year,item,crop)]['收穫面積']=globals()['df_harvest_{}_{}_{}'.format(year,item,crop)]['收穫面積'].astype(float)
            globals()['df_yield_{}_{}_{}'.format(year,item,crop)]['稻穀總產量']=globals()['df_yield_{}_{}_{}'.format(year,item,crop)]['稻穀總產量'].map(lambda x:re.sub(',','',x))
            globals()['df_yield_{}_{}_{}'.format(year,item,crop)]['稻穀總產量']=globals()['df_yield_{}_{}_{}'.format(year,item,crop)]['稻穀總產量'].astype(float)
            globals()['df_chao_{}_{}_{}'.format(year,item,crop)]['糙米總產量']=globals()['df_chao_{}_{}_{}'.format(year,item,crop)]['糙米總產量'].map(lambda x:re.sub(',','',x))
            globals()['df_chao_{}_{}_{}'.format(year,item,crop)]['糙米總產量']=globals()['df_chao_{}_{}_{}'.format(year,item,crop)]['糙米總產量'].astype(float)
```
