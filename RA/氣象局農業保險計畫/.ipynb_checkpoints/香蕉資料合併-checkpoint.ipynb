# 表格合併

```python
df1.drop(len(df1)-1,axis=0,inplace=True)
df1['年度'] = df1['年度'].map(lambda x:int(str(x).strip('0').strip('.')))
df1['災  害\n起始日'] = df1['災  害\n起始日'].map(lambda x:int(str(x).strip('0').strip('.')))
df1['颱風日期'] = df1.apply(lambda x:datetime.datetime.strptime(str(x['年度']+1911)+str(x['災  害\n起始日'])[-4:], "%Y%m%d").date(),axis=1)
```

```python
df2 = pd.melt(df2, id_vars=['縣市','鄉鎮區'])
df2.columns = ['縣市','鄉鎮','年度','種植面積(公頃)']
df2['年度'] = df2['年度'].map(lambda x:re.sub('年','',x))
```

```python
df3 = pd.melt(df3, id_vars=['縣市','鄉鎮區'])
df3.columns = ['縣市','鄉鎮','年度','產量(公斤)']
df3['年度'] = df3['年度'].map(lambda x:re.sub('年','',x))
df2['年度'] = df2['年度'].astype(int)
df3['年度'] = df3['年度'].astype(int)
```

```python
df4 = pd.merge(df2,df3,left_on=['縣市','鄉鎮','年度'],right_on=['縣市','鄉鎮','年度'])
df1 = pd.merge(df1,df4,left_on=['縣市','鄉鎮','年度'],right_on=['縣市','鄉鎮','年度'])
```

```python
for i in df1.columns:
    df1.rename(columns={'{}'.format(i):'{}'.format(re.sub('\s+','',i))},inplace=True)
```

```python
df1['不一致(面積&產量)'] = df1.apply(lambda x:1 if (x['被害面積']>x['種植面積(公頃)'] or x['損害數目(公斤)']>x['產量(公斤)']) else 0,axis=1)
df1['不一致(產量)'] = df1.apply(lambda x:1 if x['損害數目(公斤)']>x['產量(公斤)'] else 0,axis=1)
```

```python
print('災損及產量資料不一致之比例為',df1[df1['不一致(面積&產量)']==1].shape[0]/df1.shape[0])
print('災損及產量資料不一致之比例為',df1[df1['不一致(產量)']==1].shape[0]/df1.shape[0])
```
