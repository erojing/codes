```python
#檢查missing value比例
check_1 = pd.read_excel(u'98-109稻米資料.xlsx',sheet_name='一期種植面積(公頃)',index_col=0)
check_2 = pd.read_excel(u'98-109稻米資料.xlsx',sheet_name='二期種植面積(公頃)',index_col=0)
print('1期原始樣本數為：',check_1.shape[0],'\nmissing value全捨棄後樣本數為：',check_1.dropna().shape[0])
print('\n2期原始樣本數為：',check_2.shape[0],'\nmissing value全捨棄後樣本數為：',check_2.dropna().shape[0])
```

```python
#檢查missing value分布
check_1_df = pd.DataFrame()
check_1_df['縣市鄉鎮名稱'] = check_1.T.isnull().sum().index
check_1_df['na數量'] = check_1.T.isnull().sum().values
check_1_df['縣市'] = check_1_df['縣市鄉鎮名稱'].map(lambda x:x[0:3])
check_1_df['鄉鎮區'] = check_1_df['縣市鄉鎮名稱'].map(lambda x:x[3:])
check_1_df = check_1_df[check_1_df['na數量']!=0]
check_1_final = pd.DataFrame()
check_1_final['縣市'] = check_1_df.groupby('縣市')['鄉鎮區'].count().index
check_1_final['na數量'] = check_1_df.groupby('縣市')['鄉鎮區'].count().values
for i in sorted(list(check_1_df['na數量'].unique())):
    temp = pd.DataFrame()
    temp['縣市'] = check_1_df[check_1_df['na數量']==i].groupby('縣市')['鄉鎮區'].count().index
    temp['na數{}'.format(i)] = check_1_df[check_1_df['na數量']==i].groupby('縣市')['鄉鎮區'].count().values
    check_1_final = pd.merge(check_1_final,temp,on = '縣市',how = 'left')
check_1_final.fillna(0,inplace=True)
```
