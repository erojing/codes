# distance to the nearest major seaports (Shanghai, Hong Kong, and Tianjin)

```python
import requests
import json
import urllib
from urllib.request import urlopen
def getGeoForAddress(address):
    add = urllib.parse.quote(address)
    addressUrl = "https://maps.googleapis.com/maps/api/geocode/json?address=" + add + '&key='
    res = requests.get(addressUrl)
    js = json.loads(res.text)
    if js.get('results'):
        lat = js.get('results')[0]['geometry']['location']['lat']
        lng = js.get('results')[0]['geometry']['location']['lng']
    else:
        lat = ''
        lng = ''
    return [lat, lng]
```

```python
city_code = pd.read_excel('C:/Users/user/Dropbox/paper/科技園區/China/省地縣碼對照.xls')
city_code = city_code[~city_code['代碼'].isnull()].drop(['标注','说明'],axis=1)
city_code['代碼_str'] = city_code['代碼'].apply(lambda x:str(int(x)))
city_code = city_code.drop('代碼',axis=1)
```

```python
city_code.fillna('',inplace=True)
city_code['區域地址'] = city_code.apply(lambda x:x['省份']+x['市級']+x['名稱'],axis=1)
city_code.loc[:,'lat'] = ''
city_code.loc[:,'lnt'] = ''
```

```python
for i in city_code.index:
    address = city_code.loc[i,'區域地址']
    g = getGeoForAddress(address)
    city_code.loc[i,'lat'] = g[0]
    city_code.loc[i,'lnt'] = g[1]
```

```python
city_code = city_code[city_code['lat']!='']
```

```python
tianjin = (38.9832,117.7428)
hongkong = (22.2897,114.1523)
shanhai = (30.626539,122.064958)
```

```python
from geopy.distance import geodesic
def how_far(a,b):
    try:
        return geodesic(a,b).kilometers
    except:
        return np.nan
city_code['天津'] = city_code.apply(lambda x:how_far(tianjin,(x['lat'],x['lnt'])),axis=1)
city_code['香港'] = city_code.apply(lambda x:how_far(hongkong,(x['lat'],x['lnt'])),axis=1)
city_code['上海'] = city_code.apply(lambda x:how_far(shanhai,(x['lat'],x['lnt'])),axis=1)
```

```python
city_code['nearest seaport'] = city_code.apply(lambda x:np.min([x['天津'],x['香港'],x['上海']]),axis=1)
```
